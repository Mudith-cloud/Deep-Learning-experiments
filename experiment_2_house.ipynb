# Experiment 2
ANN for House Price Prediction

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import pandas as pd
```

```python
# Load dataset (replace with your dataset file name)
data = pd.read_csv('housing.csv')
data.head()
```

```python
# Assuming last column is target (price)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
# Build ANN with 1 hidden layer
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=(X.shape[1],)),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train model
model.fit(X, y, epochs=100, verbose=1)
```

```python
# Predict example
prediction = model.predict(X[:1])
print('Predicted price:', prediction[0][0])
```
