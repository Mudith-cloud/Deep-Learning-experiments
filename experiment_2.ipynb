# Experiment 2
ANN to convert Fahrenheit to Celsius

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
# Dataset: Fahrenheit to Celsius
fahrenheit = np.array([-40, -10, 0, 32, 50, 68, 77, 100, 212], dtype=float)
celsius = (fahrenheit - 32) * 5/9
```

```python
# ANN model with 1 hidden layer
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=[1]),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train model
model.fit(fahrenheit, celsius, epochs=500, verbose=0)
```

```python
# Test prediction
test_f = 100
predicted_c = model.predict([test_f])
print(f"{test_f} Fahrenheit = {predicted_c[0][0]:.2f} Celsius")
```
