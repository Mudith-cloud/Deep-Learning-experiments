```python
# Experiment 3: Housing Price Prediction using CNN (on structured data reshaped)import tensorflow as tffrom tensorflow.keras import layers, modelsimport numpy as npimport matplotlib.pyplot as plt# ===== LOAD DATA (Using Boston Housing dataset) =====from tensorflow.keras.datasets import boston_housing(x_train, y_train), (x_test, y_test) = boston_housing.load_data()# Normalize datamean = x_train.mean(axis=0)std = x_train.std(axis=0)x_train = (x_train - mean) / stdx_test = (x_test - mean) / std# Reshape data to use CNN (samples, features → pretend 1D image)x_train = x_train[..., np.newaxis]x_test = x_test[..., np.newaxis]# ===== BUILD CNN MODEL =====model = models.Sequential([    layers.Conv1D(32, 3, activation='relu', input_shape=(x_train.shape[1],1)),    layers.MaxPooling1D(2),    layers.Conv1D(64, 3, activation='relu'),    layers.Flatten(),    layers.Dense(64, activation='relu'),    layers.Dense(1)  # regression output])# Compile modelmodel.compile(optimizer='adam',              loss='mse',              metrics=['mae'])# ===== TRAIN =====history = model.fit(x_train, y_train, epochs=50, validation_data=(x_test, y_test))# ===== EVALUATE =====loss, mae = model.evaluate(x_test, y_test)print("Mean Absolute Error:", mae)# ===== PLOTS =====plt.figure()plt.plot(history.history['mae'])plt.plot(history.history['val_mae'])plt.title("MAE")plt.xlabel("Epoch")plt.ylabel("Error")plt.legend(["Train","Val"])plt.show()plt.figure()plt.plot(history.history['loss'])plt.plot(history.history['val_loss'])plt.title("Loss")plt.xlabel("Epoch")plt.ylabel("Loss")plt.legend(["Train","Val"])plt.show()
```
