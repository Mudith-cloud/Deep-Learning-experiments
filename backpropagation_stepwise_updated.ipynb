# Experiment 4

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    return x*(1-x)

x = np.array([1,0])
target = np.array([1,0])

w1,w2,w3,w4 = 0.2,-0.3,0.4,0.1
w5,w6,w7,w8 = -0.5,0.2,0.3,0.6

b1,b2 = 0.1,-0.2
b3,b4 = 0.2,-0.1

lr = 0.1
```

## 🔹 Forward Pass (Hidden Layer Computation)
Compute i1, i2 and activations a1, a2

```python
# Hidden layer forward pass
i1 = x[0]*w1 + x[1]*w3 + b1
i2 = x[0]*w2 + x[1]*w4 + b2

a1 = sigmoid(i1)
a2 = sigmoid(i2)

print(i1, i2)
print(a1, a2)
```

## Q1 & Q2: Compute i3, i4 and outputs y1, y2

```python
# Forward pass
i1 = x[0]*w1 + x[1]*w3 + b1
i2 = x[0]*w2 + x[1]*w4 + b2

a1 = sigmoid(i1)
a2 = sigmoid(i2)

i3 = a1*w5 + a2*w7 + b3
i4 = a1*w6 + a2*w8 + b4

y1 = sigmoid(i3)
y2 = sigmoid(i4)

print(i3, i4)
print(y1, y2)
```

## Q3: Compute Loss

```python
loss = 0.5*((target[0]-y1)**2 + (target[1]-y2)**2)
print(loss)
```

## Q4: Output layer errors

```python
err1 = -(target[0]-y1)*sigmoid_derivative(y1)
err2 = -(target[1]-y2)*sigmoid_derivative(y2)

print(err1, err2)
```

## Q5: Gradients for w5, w6, w7, w8

```python
dw5 = err1*a1
dw6 = err2*a1
dw7 = err1*a2
dw8 = err2*a2

print(dw5, dw6, dw7, dw8)
```

## Q6: Hidden layer errors

```python
err_h1 = sigmoid_derivative(a1)*(err1*w5 + err2*w6)
err_h2 = sigmoid_derivative(a2)*(err1*w7 + err2*w8)

print(err_h1, err_h2)
```

## Q7: Update weights

```python
w1 -= lr*err_h1*x[0]
w2 -= lr*err_h2*x[0]
w3 -= lr*err_h1*x[1]
w4 -= lr*err_h2*x[1]

w5 -= lr*dw5
w6 -= lr*dw6
w7 -= lr*dw7
w8 -= lr*dw8

b1 -= lr*err_h1
b2 -= lr*err_h2
b3 -= lr*err1
b4 -= lr*err2

print(w1,w2,w3,w4,w5,w6,w7,w8)
```

## Q8: Multiple epochs

```python
for epoch in range(50):
    i1 = x[0]*w1 + x[1]*w3 + b1
    i2 = x[0]*w2 + x[1]*w4 + b2
    a1 = sigmoid(i1)
    a2 = sigmoid(i2)
    i3 = a1*w5 + a2*w7 + b3
    i4 = a1*w6 + a2*w8 + b4
    y1 = sigmoid(i3)
    y2 = sigmoid(i4)
    loss = 0.5*((target[0]-y1)**2 + (target[1]-y2)**2)

print(loss)
```
