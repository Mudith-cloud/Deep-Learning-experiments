```python
# CIFAR-10 CNNimport tensorflow as tffrom tensorflow.keras import datasets, layers, modelsimport matplotlib.pyplot as plt(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()x_train, x_test = x_train/255.0, x_test/255.0plt.figure(figsize=(10,2))for i in range(5):    plt.subplot(1,5,i+1)    plt.imshow(x_train[i])    plt.axis('off')plt.show()model = models.Sequential([    layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),    layers.MaxPooling2D(2,2),    layers.Conv2D(64,(3,3),activation='relu'),    layers.Flatten(),    layers.Dense(64,activation='relu'),    layers.Dense(10)])model.compile(optimizer='adam',              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),              metrics=['accuracy'])history = model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))plt.plot(history.history['accuracy'])plt.plot(history.history['val_accuracy'])plt.show()
```
